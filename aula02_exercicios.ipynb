## <b>Inteiros (int)</b>

### 1. Escreva um programa que soma dois números inteiros inseridos pelo usuário.

```python
def soma_inteiros(num1, num2):
    return num1 + num2
```

```python
num1 = int(input("Digite o primeiro número inteiro: "))
print(num1)
```

```python
num2 = int(input("Digite o segundo número inteiro: "))
print(num2) 
```

```python
print(f"Resultado da soma: {soma_inteiros(num1, num2)}")
```

### 2. Crie um programa que receba um número do usuário e calcule o resto da divisão desse número por 5.

```python
def resto_divisao_por_5(num):
    return num % 5
```

```python
dividendo = int(input("Digite o dividendo: "))
print(dividendo)
```

## <b>Números de Ponto Flutuante (float) </b>

### 6. Escreva um programa que receba dois números flutuantes e realize sua adição.

```python
def soma_float(num1, num2):
    return num1 + num2
```

```python
num1 = float(input("Digite o primeiro número flutuante: "))
print(num1)
```

```python
num2 = float(input("Digite o segundo número flutuante: "))
print(num2)
```

```python
print(f"Restultado da soma entre {num1} e {num2} é {soma_float(num1, num2)}")
```

### 7. Crie um programa que calcule a média de dois números flutuantes fornecidos pelo usuário.

### 8. Desenvolva um programa que calcule a potência de um número (base e expoente fornecidos pelo usuário).

```python
def potencia(base, expoente):
    return base ** expoente
```

```python
base = float(input("Digite a base: "))
print(base)
```

```python
expoente = float(input("Digite o expoente: "))
print(expoente)
```

```python
print(f"O resultado da potência é {potencia(base, expoente)}")
```

### 9. Faça um programa que converta a temperatura de Celsius para Fahrenheit.

```python
def celsius_fahrenheit(temp_celsius):
    return (temp_celsius * (9/5)) + 32
```

```python
temp_celsius = float(input("Digite a temperatura em Celsius: "))
print(temp_celsius)
```

```python
print(f"{temp_celsius}°C é igual a {celsius_fahrenheit(temp_celsius):.2f}°F")
```

### 10. Escreva um programa que calcule a área de um círculo, recebendo o raio como entrada.

```python
PI = 3.14159
```

```python

def area_circulo(raio):
    return PI * raio ** 2
```

```python
raio = float(input("Digite o raio do círculo: "))
print(raio)
```

```python
print(f"A área do círculo de raio {raio} é igual a {area_circulo(raio):.2f}")
```
